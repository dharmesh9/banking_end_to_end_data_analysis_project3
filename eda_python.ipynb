```python
parts = [0,100000,300000,float('inf')]
lables = ['Low', 'Medium', 'High']

df['Income Band'] = pd.cut(df['Estimated Income'], bins=parts, labels=lables, right = False)
```

```python
df['Income Band'].value_counts()
```

```python
df['Income Band'].value_counts().plot(kind='barh')
```

```python
categorical_col = [
    'Nationality',
    'Occupation',
    'Fee Structure',
    'Loyalty Classification',
    'Properties Owned',
    'Risk Weighting',
    'BRId',
    'GenderId',
    'IAId',
    'Income Band'
]

for col in categorical_col:
    print(f"Value Counts for '{col}':")
    print(df[col].value_counts())
```

```python
for i,predictor in enumerate(df[[
    'Nationality',
    'Occupation',
    'Fee Structure',
    'Loyalty Classification',
    'Properties Owned',
    'Risk Weighting',
    'BRId',
    'GenderId',
    'IAId',
    'Income Band'
]]):
        
    plt.figure(i)
    sns.countplot(data=df,x=predictor)
```
