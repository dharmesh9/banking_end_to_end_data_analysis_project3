```python
parts = [0,100000,300000,float('inf')]
lables = ['Low', 'Medium', 'High']

df['Income Band'] = pd.cut(df['Estimated_Income'], bins=parts, labels=lables, right = False)
```

```python
df['Income Band'].value_counts()
```

```python
df['Income Band'].value_counts().plot(kind='barh')
```

```python
categorical_col = [
    'Nationality',
    'Occupation',
    'Fee_Structure',
    'Loyalty_Classification',
    'Properties_Owned',
    'Risk_Weighting',
    'BRId',
    'GenderId',
    'IAId',
    'Income Band'
]

for col in categorical_col:
    print(f"Value Counts for '{col}':")
    print(df[col].value_counts())
```

```python
for i,predictor in enumerate(df[[
    'Nationality',
    'Occupation',
    'Fee_Structure',
    'Loyalty_Classification',
    'Properties_Owned',
    'Risk_Weighting',
    'BRId',
    'GenderId',
    'IAId',
    'Income Band'
]]):
        
    plt.figure(i)
    sns.countplot(data=df,x=predictor)
```

```python
for i,predictor in enumerate(df[[
    'Nationality',
    'Occupation',
    'Fee_Structure',
    'Loyalty_Classification',
    'Properties_Owned',
    'Risk_Weighting',
    'BRId',
    'GenderId',
    'IAId',
    'Income Band'
]]):
        
    plt.figure(i)
    sns.countplot(data=df,x=predictor,hue='GenderId')
```

```python
for i,predictor in enumerate(df[[
    'Nationality',
    'Occupation',
    'Fee_Structure',
    'Loyalty_Classification',
    'Properties_Owned',
    'Risk_Weighting',
    'BRId',
    'GenderId',
    'IAId',
    'Income Band'
]]):
        
    plt.figure(i)
    sns.countplot(data=df,x=predictor,hue='Nationality')
```

```python
num_cols = df.select_dtypes(include='number').columns
df[num_cols].hist(figsize=(15,12), bins=20)
```

```python
for col in num_cols:
    plt.figure(figsize=(6,3))
    sns.boxplot(df[col])
    plt.title(col)
    plt.show()
```

```python
for col in categorical_col:
    if col == 'Occupation':
        continue
    plt.figure(figsize=(8,6))
    sns.histplot(df[col])
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.title('Histrogram of Occupation Counts')
    plt.show()
```
