```python
parts = [0,100000,300000,float('inf')]
lables = ['Low', 'Medium', 'High']

df['Income Band'] = pd.cut(df['Estimated_Income'], bins=parts, labels=lables, right = False)
```

```python
df['Income Band'].value_counts()
```

```python
df['Income Band'].value_counts().plot(kind='barh')
```

```python

categorical_col = [
    'Nationality',
    'Occupation',
    'Fee_Structure',
    'Loyalty_Classification',
    'Properties_Owned',
    'Risk_Weighting',
    'BRId',
    'GenderId',
    'IAId',
    'Income Band'
]

for col in categorical_col:
    print(f"Value Counts for '{col}':")
    print(df[col].value_counts())
```

```python
# Visualizations
for i,predictor in enumerate(df[[
    'Nationality',
    'Occupation',
    'Fee_Structure',
    'Loyalty_Classification',
    'Properties_Owned',
    'Risk_Weighting',
    'BRId',
    'GenderId',
    'IAId',
    'Income Band'
]]):
        
    plt.figure(i)
    sns.countplot(data=df,x=predictor)
```

```python
for i,predictor in enumerate(df[[
    'Nationality',
    'Occupation',
    'Fee_Structure',
    'Loyalty_Classification',
    'Properties_Owned',
    'Risk_Weighting',
    'BRId',
    'GenderId',
    'IAId',
    'Income Band'
]]):
        
    plt.figure(i)
    sns.countplot(data=df,x=predictor,hue='GenderId')
```

```python
num_cols = df.select_dtypes(include='number').columns
df[num_cols].hist(figsize=(15,12), bins=20)
```

```python
for col in num_cols:
    plt.figure(figsize=(6,3))
    sns.boxplot(df[col])
    plt.title(col)
    plt.show()
```

```python
for col in categorical_col:
    if col == 'Occupation':
        continue
    plt.figure(figsize=(8,6))
    sns.histplot(df[col])
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.title('Histogram of Occupation Counts')
    plt.show()
```

```python
numerical_col = ['Estimated_Income', 'Superannuation_Savings','Credit_Card_Balance','Bank_Loans','Bank_Deposits','Checking_Accounts','Saving_Accounts','Foreign_Currency_Account','Business_Lending']
plt.figure(figsize=(15,8))
for i, col in enumerate(numerical_col):
    plt.subplot(3,3,i+1)
    sns.histplot(df[col], kde=True)
    plt.title(col)
plt.show()
```

```python
numerical_col = ['Estimated_Income', 'Superannuation_Savings','Credit_Card_Balance','Bank_Loans','Bank_Deposits','Checking_Accounts','Saving_Accounts','Foreign_Currency_Account','Business_Lending']
 
correlation_matrix = df[numerical_col].corr()
plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot=True,cmap='crest',fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Income Band', y='Bank_Deposits', data=df)
plt.title('Deposits by Income Band')
plt.show()
```

```python
sns.pairplot(df[numerical_col])
plt.show()
```

```python
for col in numerical_col:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    
    outliers = df[(df[col] < q1 - 1.5*iqr) | (df[col] > q3 + 1.5*iqr)]
    print(f"{col}: {len(outliers)} outliers")
```

```python
df[numerical_col].skew()
```

```python
df['debt_to_income'] = df['Bank_Loans'] / df['Estimated_Income']
df['savings_ratio'] = df['Saving_Accounts'] / df['Estimated_Income']
```
